```python
import csv
import numpy as np
import pandas as pd
```

```python
csv_path = "data/NFLX.csv"
df = pd.read_csv(csv_path, usecols=range(0,5))
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.query('2021 <= Date.dt.year')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Date'], df['Close'])
plt.title('NFLX Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.show
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month

df1 = df.groupby(['year', 'month']).mean().reset_index()

plt.plot(df1['Date'], df1['Close'])
plt.title('High, Low, and Close Prices since January 2002')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show
```

```python
df2 = df.query('2022 <= Date.dt.year')

plt.plot(df2['Date'], df2['Close'], label='Close')
plt.plot(df2['Date'], df2['Low'], label='Low')
plt.plot(df2['Date'], df2['High'], label='High')
plt.legend()
plt.title('Monthly Average Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show
```
